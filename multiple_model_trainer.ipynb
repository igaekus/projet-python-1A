# Multiple model trainer

This notebook is expected to generalize simple model trainer for several stocks and to plot them in a suitable way



```python
%run stock_dataset_creator.ipynb
%run simple_model_trainer.ipynb
from tqdm import tqdm
```

```python
def multiple_model_trainer(dico):
    n,df,l = len(dico),{},list(dico.keys())
    for i in tqdm(range(0,n)):
        df[l[i]] = simple_model_trainer(dico[l[i]], lagmax=13)
    
    if n%2!=0 : rows = n+1
    else : rows = n
    fig, ax = plt.subplots(nrows=int(rows/2), ncols=2, figsize=(16, 9),constrained_layout=False)
    axes = ax.flatten()
    i=0
    for key in dico.keys():
        df[key].plot(title="Models predicitons for "+key+" stocks", ax=axes[i])
        axes[i].grid(True)
        #plt.axvline(label = "Model starts forecasting" , x=X_test_end, ymin=0, ymax=190, lw = 4, color = "red")
        i+=1
    plt.tight_layout()
    plt.legend()
    plt.show()
    
        
        
```

```python
stocks = {"loreal" : "OR.PA", "google" : "GOOGL", "credit_agricole" : "ACA.PA", "cac40":"^FCHI", "APPLE": "AAPL", "Sogé":"GLE.PA"}
```

```python
multiple_model_trainer(stocks)
```
