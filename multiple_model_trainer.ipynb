```python
def multiple_model_trainer(dico):
    df = {key : simple_model_trainer(dico[key]) for key in dico.keys()}
    n = len(dico)
    if n%2!=0 : rows = n+1
    else : rows = n
    fig, ax = plt.subplots(nrows=int(rows/2), ncols=2, figsize=(16, 9),constrained_layout=True)
    axes = ax.flatten()
    i=0
    for key in dico.keys():
        df[key].plot(title="Models predicitons for "+key+" stocks", ax=axes[i])
        axes[i].grid(True)
        #plt.axvline(label = "Model starts forecasting" , x=X_test_end, ymin=0, ymax=190, lw = 4, color = "red")
        i+=1
    plt.tight_layout()
    plt.legend()
    plt.show()
    
        
        
```
