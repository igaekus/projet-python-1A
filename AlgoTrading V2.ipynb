```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf

name = "AAPL"
start, end = "2012-01-01", "2019-01-01"
stock_df = pd.DataFrame(yf.Ticker(name).history(period="max"))[start: end]
stock_df.drop(["High", "Low", "Open", 'Dividends', "Stock Splits"], axis = 1, inplace = True)
stock_df["daily log returns"] = np.log(stock_df["Close"]/stock_df["Close"].shift(1))
stock_df.dropna()

```
